17.	Identify the two features with the highest positive correlation and the two with the highest negative correlation. 

```python
import pandas as pd
df = pd.read_csv('spambase(2).csv')
corr_matrix = df.corr()

# Find the highest positive correlation
max_pos_corr = correlation_matrix.unstack().sort_values(ascending=False)
max_pos_corr = max_pos_corr[max_pos_corr < 1].head(1)  # Exclude self-correlations

# Find the highest negative correlation
max_neg_corr = correlation_matrix.unstack().sort_values(ascending=True)
max_neg_corr = max_neg_corr.head(1)

print("Highest positive correlation:", max_pos_corr)
print("Highest negative correlation:", max_neg_corr)
```

18.	Visualize the correlation matrix using a heatmap. 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('spambase(2).csv')
numeric_df = df.select_dtypes(include='number')
corr_matrix = numeric_df.corr()
plt.figure(figsize=(50, 50))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```
